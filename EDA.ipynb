## âš”ï¸Challenge Loop through all the columns and get the outliers and replace with NaN for each

```python

def replace_outliers_with_nan(data_table, columns_list):
    cleaned_df = data_table.copy()  # Copy the original DataFrame

    for column in columns_list:
        q1 = cleaned_df[column].quantile(0.25)  # First quartile (25th percentile)
        q3 = cleaned_df[column].quantile(0.75)  # Third quartile (75th percentile)
        IQR = q3 - q1  # Interquartile range

        lower_bound = q1 - 1.5 * IQR  # Lower bound
        upper_bound = q3 + 1.5 * IQR  # Upper bound

        # Replacing only outliers with NaN in that specific column, keeping the rest of the row
        cleaned_df[column] = cleaned_df[column].mask((cleaned_df[column] < lower_bound) | (cleaned_df[column] > upper_bound), None)

    return cleaned_df
```

```python
columns = df.select_dtypes(include=['number']).columns.tolist()
columns
```

```python
newDF = replace_outliers_with_nan(df,columns)
```

## ðŸ§ªExperiment with mask() and loc[]

* When replacing outliers in a dataset, it is important to ensure that only the outlier values are replaced with NaN, while the rest of the row remains intact.
* âš ï¸ Issue with .loc[]
Using .loc[] incorrectly can result in replacing entire rows with NaN, which is not what we want
* To ensure that only the outlier values are replaced, we use .mask().


```python
data = {'A': [1, 2, 4, 4, 5],
        'B': [3, 4, 3, 4, 1],
        'C': [5, 6, 7, 8, 9],
        'D': [1, 2, 3, 1, 5]}

data = pd.DataFrame(data)
```

```python
data
```

```python
data.columns
```

```python
for col in data.columns:
    data.loc[(data[col] > 0) & (data[col] < 2)] = None
```

## âš ï¸ Multicolinearity in Independent Variables troublesome to our models

ðŸ” What is Multicollinearity?
Multicollinearity occurs when two or more independent variables are highly correlated in a dataset.

ðŸ“Œ Key takeaway: Instead of each variable bringing unique information, they overlap, making it difficult for a model to determine which one truly impacts the target variable.

## ðŸ› ï¸ How to Detect Multicollinearity?

## ðŸ“Œ Variance Inflation Factor (VIF)
A variance inflation factor (VIF) is a measure of the amount of multicollinearity in regression analysis. Multicollinearity exists when there is a correlation between multiple independent variables in a multiple regression model. This can adversely affect the regression results. Thus, the variance inflation factor can estimate how much the variance of a regression coefficient is inflated due to multicollinearity.

* VIF = 1 â†’ No multicollinearity
* VIF > 5 â†’ Moderate multicollinearity
* VIF > 10 â†’ Severe multicollinearity (must be addressed!)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X = newDF[['mean perimeter', 'mean radius', 'worst area']]
```

```python

X = X.apply(lambda col: col.fillna(0)) # as the stats model dont work on NaN or Inf values
vif_data = pd.DataFrame()
vif_data['Features'] = X.columns
vif_data['VIF Factor'] = [variance_inflation_factor(X.values,i) for i in range(len(X.columns))]
vif_data
```
